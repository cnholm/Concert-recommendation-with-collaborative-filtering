```python
import pandas as pd
%matplotlib inline
```

```python
ratings = pd.read_csv('tribute_data.csv', sep='\t',
                      names=['user', 'item', 'rating', 'timestamp'])
ratings.head()
```

```python
# Select duplicate rows except first occurrence based on all columns
duplicateRowsDF = ratings[ratings.duplicated(keep=False)]
 
print("Duplicate Rows except first occurrence based on all columns are :")
print(duplicateRowsDF)
```

```python
# Select all duplicate rows based on two columns
duplicateRowsDF = ratings[ratings.duplicated(['user', 'item'], keep=False)]
 
print("Duplicate Rows based on two columns are:", duplicateRowsDF, sep='\n')
```

```python
# Drop rows where all data is the same
ratings_no_dup = ratings.drop_duplicates(['user', 'item'])
print(ratings_no_dup)
```

```python
export_csv = ratings_no_dup.to_csv (r'tribute_data_no_dup.csv', index = None, header=False, sep='\t')
```
